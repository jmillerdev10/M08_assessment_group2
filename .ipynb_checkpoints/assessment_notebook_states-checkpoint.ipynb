```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# pd.Dataframe(r'https://api.census.gov/data/2018/abscs?get=NAME,GEO_ID,NAICS2017_LABEL,SEX,ETH_GROUP,RACE_GROUP,VET_GROUP,FIRMPDEMP&for=us:*&NAICS2017=00&key=bccfff37a66ca0c564d5d5d80d0236771b334f9b',header='False')
# df.to_json(orient='table')
company = pd.read_json(r'https://api.census.gov/data/2018/abscs?get=NAME,FIRMPDEMP&for=state&NAICS2017=00&key=bccfff37a66ca0c564d5d5d80d0236771b334f9b')
company.columns = company.iloc[0]
company =  company.iloc[1:].reset_index(drop=True)
company['NAME'] = company['NAME'].astype(str)
company['FIRMPDEMP'] = company['FIRMPDEMP'].astype(int)
company

# airports['elevation_ft'] = airports['elevation_ft'].astype(int)
```

```python
company = company[['NAME','FIRMPDEMP','state']]
company
```

```python
tech = pd.read_json(r'https://api.census.gov/data/2018/abstcb?get=NAME,GEO_ID,TECHUSE,IMPACTWF_U,FIRMPDEMP,EMP,PAYANN&for=state&NAICS2017_LABEL&key=bccfff37a66ca0c564d5d5d80d0236771b334f9b')
tech.columns = tech.iloc[0]
tech =  tech.iloc[1:].reset_index(drop=True)
tech['FIRMPDEMP'] = tech['FIRMPDEMP'].astype(int)
tech['EMP'] = tech['EMP'].astype(int)
tech['PAYANN'] = tech['PAYANN'].astype(int)

tech
```

```python
business = pd.read_json(r'http://api.census.gov/data/2018/abscb?get=BUSCHAR_LABEL,QDESC,QDESC_LABEL,FIRMPDEMP,EMP&for=state&BUSCHAR=BM&BUSCHAR=BN&key=bccfff37a66ca0c564d5d5d80d0236771b334f9b')

business.columns = business.iloc[0]

business = business.iloc[1:].reset_index(drop=True)

business.rename(columns={'FIRMPDEMP':'FIRM_COUNT'}, inplace=True)

business['FIRM_COUNT'] = business['FIRM_COUNT'].astype(int)
business['EMP'] = business['EMP'].astype(int)

business
```

```python
business['avg_emp_per_firm'] = business['EMP'] / business['FIRM_COUNT']
business
```

```python
business[business['BUSCHAR'] == 'BM']
# bus_fam = business[business['BUSCHAR'] == 'BM']
# bus_nonfam = business[business['BUSCHAR'] == 'BN']

# bus_fam.rename(columns={'FIRM_COUNT':'FAMILY_OWNED_FIRMS','EMP':'FAMILY_OWNED_EMPLOYEES','avg_emp_per_firm':'avg_emp_per_family_owned_firm'}, inplace=True)
# bus_nonfam.rename(columns={'FIRM_COUNT':'NONFAMILY_OWNED_FIRMS','EMP':'NONFAMILY_OWNED_EMPLOYEES','avg_emp_per_firm':'avg_emp_per_nonfamily_owned_firm'}, inplace=True)
```

```python
cs_bc = company.merge(business, how='inner', on=["state"])
cs_bc = cs_bc[['NAME','BUSCHAR_LABEL','FIRMPDEMP','state','FIRM_COUNT','EMP','avg_emp_per_firm']]

cs_bc
```

```python
sorted_cs_bc = cs_bc.sort_values(by='NAME', ascending=True)
sorted_cs_bc
```

```python
plt.figure(figsize = (15,10))
ax = sns.barplot(x=sorted_cs_bc[0:50]['NAME'], y=sorted_cs_bc[0:50]['avg_emp_per_firm'], hue='BUSCHAR_LABEL', data=cs_bc)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
ax.set_xlabel('State', size=18)
ax.set_ylabel('Avg Number of Employees per Firm', size=18) 


```
